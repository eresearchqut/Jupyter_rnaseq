## 2b. Differential expression workflow overview <a class="anchor" id="overview2"></a>

### Downstream: R packages

A count table of reads per sample per gene (samples as columns, genes as rows) is generated by nfcore/rnaseq and is used as input for the downstram analysis.

The primary downstream analysis is differential expression, using the R package [DESeq2](https://bioconductor.org/packages/release/bioc/html/DESeq2.html):

> Estimate variance-mean dependence in count data from high-throughput sequencing assays and test for differential expression based on a model using the negative binomial distribution

From DESEq2 results, volcano plots, heapmaps and tables of significantly differentially expressed (DE) genes are genereated.

Other downstream analyses include:

* Outlier and batch effects

  * **Density plots** show the distribution of read density per sample. They can identify whether particular samples have a bias in the read count per gene, per sample. A strong bias can be an indication of errors in library preparation and batch effects.
  * [**PCA plots**](https://setosa.io/ev/principal-component-analysis/) show linear combinations of the ‘principal components’ in a dataset, i.e. the linearly uncorrelated variables as estimated by the R package `prcomp`. PC1 displays the majority of the variation, decreasing rapidly to PC2 and through to PCn (n = variable/sample size). Groups have been coloured and shaded based on 85% confidence levels for data range.
  * [**Dendrograms and heatmaps**](https://docs.tibco.com/pub/spotfire/7.0.1/doc/html/heat/heat_what_is_a_heat_map.htm). Dendrograms show the hierarchical clustering of samples, based on pairwise Euclidean distance. Heatmaps show, using a colour gradient, pairwise Euclidean distance between individual samples.


* Functional annotation (run in separate Notebook)

  * [KEGG pathway](https://www.genome.jp/kegg/) and [GO term](http://geneontology.org/) enrichment, based on differentially expressed genes, using the R tool [clusterProfiler](https://bioconductor.org/packages/release/bioc/html/clusterProfiler.html)
